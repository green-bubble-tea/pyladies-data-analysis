#### Getting a dominant color from image

```python
images_db = os.path.join("data", "img_db")
sample_image = Image.open(os.path.join(images_db, "1.png"))
image_array = np.array(sample_image)
```

```python
def get_dominant_color(rgb_image):
    color_index = {
        0: "red",
        1: "orange",
        2: "yellow",
        3: "green",
        4: "cyan",
        5: "blue",
        6: "pink"
    }
    hsv_image = np.array(rgb_image.convert("HSV"))
    H = hsv_image[:, :, 0] * (360 / 255)
    v = hsv_image[:, :, 2]
    
    red_min, red_max = 320, 20
    orange_min, orange_max = 21, 49
    yellow_min, yellow_max = 50, 75
    green_min, green_max = 76, 150
    cyan_min, cyan_max = 151, 199
    blue_min, blue_max = 200, 260
    pink_min, pink_max = 261, 319
    
#     col = [not(np.isin(H, range(20, 321))),
#            np.isin(H, range(21,50)), 
#            np.where(np.isin(H, range(50, 76))), 
#            np.where(np.isin(H, range(76, 151))), 
#            np.where(np.isin(H, range(151, 200))),
#            np.where(np.isin(H, range(200, 261))), 
#            np.where(np.isin(H, range(261, 320)))]
#     print(col)
 
    colors = [i[0].size for i in 
              [np.where(np.logical_and(v > 0, np.logical_or(H >= red_min, H <= red_max))),
              np.where(np.logical_and(H >= orange_min, H <= orange_max)),
              np.where(np.logical_and(H >= yellow_min, H <= yellow_max)),
              np.where(np.logical_and(H >= green_min, H <= green_max)),
              np.where(np.logical_and(H >= cyan_min, H <= cyan_max)),
              np.where(np.logical_and(H >= blue_min, H <= blue_max)),
              np.where(np.logical_and(H >= pink_min, H <= pink_max))]]
 
    max_color_index = color_index.get(colors.index(max(colors)))
    return max_color_index
    
pokemon_workdata["color"] = ""
for index, row in pokemon_workdata.iterrows():
    pokedex_path = f'{int(row["pokedex_number"])}.png'
    pokemon_img = Image.open(os.path.join(images_db, pokedex_path))
    color = get_dominant_color(pokemon_img)
    pokemon_workdata.loc[index, "color"] = color
pokemon_workdata.head()
    
```

#### Aggregating functions

How many green pokemons weight more than the average? 

```python
avg_weight = pokemon_workdata["weight_kg"].mean()
poks = pokemon_workdata.where(np.logical_and(pokemon_workdata["weight_kg"] > avg_weight, pokemon_workdata["color"] == "green"))
f = poks[~poks.isnull().any(axis=1)]
f.shape[0]
```

Visualize 

```python
mean_by_color = pokemon_workdata.groupby("color").mean().sort_values(by="attack")
mean_by_color.plot(kind="bar", y="attack", color=mean_by_color.index)
```
